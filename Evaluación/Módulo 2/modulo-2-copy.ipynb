En este módulo, introduciremos los conceptos estadísticos más básicos, que nos permitirán analizar de manera más profunda nuestros datos. 

```python
import pandas as pd 
```

```python
# https://ourworldindata.org/meat-production
data = pd.read_csv("global-meat-production-by-livestock-type.csv")
data.head()
```

```python
data.columns
```

```python
data.columns[3:]
```

```python
new_names = [
    "sheep", "beef", "pig", "game", "duck", "poultry", "horse", "camel", "goose"
]
```

```python
to_substitute = {
    key:val for key, val in zip(data.columns[3:], new_names)
}
```

```python
data = data.rename(columns=to_substitute)

data.head()
```

```python
data["Entity"].unique()
```

```python
americas = data["Entity"]=="Americas"
data[americas].head()
```

```python
no_code = data["Code"].isna()
data[no_code].head()
```

```python
regions = data[no_code].groupby("Entity").count().index
regions
```

```python
not_geographic_subregions = [
    'Africa', 
    'Americas', 
    'Asia', 
    'Australia & New Zealand', 
    'Belgium-Luxembourg',
    'Ethiopia PDR',
    'Europe',
    'European Union',
    'Land Locked Developing Countries', 
    'Least Developed Countries',
    'Low Income Food Deficit Countries',
    'Net Food Importing Developing Countries',
    'Pacific Islands Trust Territory',
    'Oceania',
    'Small island developing States',
    'Sudan (former)',
    ]
```

```python
subregions = [region for region in regions if region not in not_geographic_subregions]
subregions
```

```python
data = data[data["Entity"].isin(subregions)]
data.head()
```

```python
data = data.drop(columns="Code").set_index(["Entity", "Year"])
data
```

```python
data.sum()
```

```python
data.sum().plot(kind="pie", autopct='%1.0f%%')
```

```python
data.sum(axis=0)
```

```python
data.sum(axis=1)
```

```python
data.groupby("Entity").sum()
```

```python
data.groupby("Year").sum()
```

```python
data.groupby("Entity").sum().sum(axis=1)
```

```python
data.groupby("Entity").sum().sum(axis=1).plot(kind="bar")
```

```python
data.groupby("Year").sum().sum(axis=1)
```

```python
data.groupby("Year").sum().sum(axis=1).plot()
```
