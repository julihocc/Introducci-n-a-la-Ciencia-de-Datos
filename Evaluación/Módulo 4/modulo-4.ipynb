Una tienda de deportes de San Luis tiene dos tipos de raquetas de tenis, la de Serena Williams y la marca de aut√≥grafos Maria Sharapova. 

La demanda de los consumidores de cada marca depende no s√≥lo de su propio precio, sino tambi√©n del precio de la marca  competidora.

$D_1 = 300-20x+30y$ 

raquetas al a√±o y la demanda de raquetas Sharapova ser√°

$D_2=200+40x-10y$

Exprese los ingresos anuales totales de la tienda por la venta de estas raquetas en funci√≥n de los precios $x,y$.

### Funciones de producci√≥n Cobb-Douglas

La producci√≥n Q de una f√°brica suele considerarse una funci√≥n de la cantidad K de capital inversi√≥n de capital y del tama√±o L de la mano de obra. Las funciones de producci√≥n de la forma

$Q(K,L)= AK^\alpha L^\beta$

donde $A,\alpha, \beta$ son constantes positivas con $\alpha+\beta = 1$ han demostrado ser especialmente √∫tiles en el an√°lisis econ√≥mico y son conocidas como funciones de producci√≥n Cobb-Douglas. 

### Producci√≥n como una funci√≥n de dos variables

Supongamos que en una determinada f√°brica, la producci√≥n viene dada por la funci√≥n de producci√≥n Cobb-Douglas

$Q(K,L)=60K^{1/3}L^{2/3}$

unidades, donde K es la inversi√≥n de capital medida en unidades de 1.000 d√≥lares y L el tama√±o de la mano de obra medida en horas-trabajador.

1. Calcule la producci√≥n si la inversi√≥n de capital es de 512.000 d√≥lares y se utilizan 1.000 horas-trabajo
de mano de obra.

2.  Demuestre que la producci√≥n de la parte (a) se duplicar√° si tanto la inversi√≥n de capital como el tama√±o de la mano de obra se duplican.

```python
def Q(K, L):
    return 60*K**(1/3)*L**(2/3)
```

```python
k = np.linspace(0, 1024)
l = np.linspace(0, 2000) 

K,L = np.meshgrid(k,l)

Z = Q(K, L)

plt.contour(K, L, Z)
plt.colorbar()
```

```python
K0,L0 = (512, 1000)
Q(K0, L0)
```

```python
Q(2*K0, 2*L0)
```

```python
Q(2*K0, 2*L0)/Q(K0, L0)
```

### Valor presente como una funci√≥n de cuatro variables

Recordemos que el valor actual de B d√≥lares en t a√±os invertidos a la tasa anual r compuesta k veces por a√±o viene dado por

$P(B,r,k,t)=B\left(1+\frac{r}{k}\right)^{-kt}$

Encuentre el valor presente de $\$10,000$ en 5 a√±os invertido al 6\% anual, compuesto cuatrimestralmente. 



```python
def P(B, r, k, t):
    return B*np.power(1+r/k, -k*t)
```

```python
P(10_000, 0.06, 4, 5)
```

### Poblaci√≥n como una funci√≥n de tres variables

Una poblaci√≥n que crece exponencialmente satisface

$P(A,k,t)= Ae^{kt}$

donde P es la poblaci√≥n en el momento t, A es la poblaci√≥n inicial (cuando t = 0), y k es
la tasa de crecimiento relativo (per c√°pita). 

La poblaci√≥n de un determinado pa√≠s es actualmente de
5 millones de personas y crece a un ritmo del 3% anual. ¬øCu√°l ser√° la poblaci√≥n dentro de 7 a√±os?



```python
def P(A, k, t):
    return A*np.exp(k*t)
```

```python
P(5, 0.03, 7)
```

### Curvas de nivel en Econom√≠a: isocuantas y curvas de indiferencia Curvas de indiferencia

Las curvas de nivel aparecen en muchas aplicaciones diferentes. Por ejemplo, en econom√≠a, si la producci√≥n 
$Q(x, y)$ de un proceso de producci√≥n est√° determinada por dos insumos $x,y$ (por ejemplo, horas
de trabajo e inversi√≥n de capital), la curva de nivel $Q(x, y) = C$ se denomina curva
del producto constante C o, m√°s brevemente, una isocuanta (iso significa "igual").

Otra aplicaci√≥n de las curvas de nivel en econom√≠a es el concepto de curvas de indiferencia. Un consumidor que est√° considerando la compra de varias unidades de de cada uno de los dos productos se asocia con una funci√≥n de utilidad $U(x, y)$, que mide la satisfacci√≥n total (o utilidad) que obtiene el consumidor por tener $x$ unidades del primer producto y $y$ unidades del segundo.

Una curva de nivel $U(x, y)=C$ de la funci√≥n de utilidad se llama curva de indiferencia y da todas las combinaciones de $x, y$ que conducen al mismo nivel de satisfacci√≥n del consumidor. 

### Aplicaci√≥n de las curvas de nivel a la econom√≠a

Supongamos que la utilidad que obtiene un consumidor de $x$ unidades de un bien y de $y$ unidades
de una segunda mercanc√≠a viene dada por la funci√≥n de utilidad $U(x, y)=x^{3/2}y$. 

Si el consumidor posee actualmente $x=16$ unidades de la primera mercanc√≠a y  $y=20$ unidades de la
segunda, encuentre el nivel actual de utilidad del consumidor y dibuje la correspondiente curva de indiferencia correspondiente.

```python
def U(x,y):
    return x**(3/2)*y 
```

```python
x0, y0 = (16, 20)

x = np.linspace(0.1,2*x0)
y = np.linspace(0.1,2*y0)

X,Y = np.meshgrid(x,y)
Z = U(X, Y)

y_level = U(x0,y0)/x**(3/2)

plt.xlim(0, 2*x0)
plt.ylim(0, 2*y0)
plt.plot([x0], [y0], marker="o", markersize=10)
plt.plot(x, y_level)
plt.contour(X,Y,Z, 100)
```

### Utilizaci√≥n del an√°lisis marginal para estudiar la producci√≥n

Actualmente, la plantilla est√° formada por 30 trabajadores cualificados y 60 no cualificados. 

Utilice el an√°lisis marginal para estimar el cambio en la producci√≥n semanal que resultar√° de la adici√≥n de 1 trabajador cualificado m√°s si el n√∫mero de trabajadores no cualificados no se modifica.

```python
from sympy import *
```

```python
x,y = symbols("x,y", real=True)

Q = 1200*x+500*y+x**2*y-x**3-y**2
Q 
```

```python
Qx = Q.diff(x)
Qx 
```

```python
Qx = lambdify((x,y),Qx)
Qx(30,60) 
```

```python
Q = lambdify((x,y), Q)
Q(31,60)-Q(30,60)
```

Si $Q(K, L)$ es el producto de un proceso de producci√≥n que implica el gasto de $K$ unidades de capital y $L$ unidades de trabajo, entonces la derivada parcial $Q_K(K, L)$ se denomina productividad marginal del capital y mide la tasa en cuyo producto $Q$ cambia con respecto al gasto de capital cuando la fuerza de trabajo se mantiene constante. 

De manera similar, la derivada parcial $Q_L(K, L)$ se llama productividad marginal del trabajo y mide la tasa de cambio de la producci√≥n con respecto al nivel de trabajo cuando el gasto de capital se mantiene constante.

###  La regla de la cadena para derivadas parciales

Ana Laura administra una tienda naturista que vende dos tipos de agua con vitaminas, la marca A y la marca B. Las cifras de ventas indican que si la marca A se vende a x d√≥lares por botella y la marca B a y d√≥lares por botella, la demanda de la marca A ser√°

$Q(x,y)=300-20x^2+30y$ botellas por mes

Ella estima que aqu√≠ a $t$ meses el precio de la marca A sera

$x = 2+0.05t$ d√≥lares por bote

y el precio de la marca B ser√° 

$y = 2+0.1\sqrt{t}$ d√≥lares por bote.

¬øA que ritmo deber√≠a Ana Laura esperar que la demanda para la marca A  cambie con respecto del tiempo de aqu√≠ a 4 meses? ¬øDeber√≠a esperar que la demanda se incremente o decrezca en ese tiempo?

```python
x,y,t = symbols("x,y,t", nonnegative=True)
t
```

```python
f = 2+0.05*t 
f 
```

```python
change = {x:f,y:g}
F = Q.subs(change) 
F
```

### Aproximaci√≥n incremental de la producci√≥n

En cierta f√°brica, la producci√≥n diaria es $Q = 60K^{1/2}L^{1/3}$ unidades, donde K denota la inversi√≥n de capital medida en unidades de $\$1,000$ y L el tama√±o de la fuerza laboral medida en horas-trabajador. La inversi√≥n de capital actual es de $\$900,000$ y 1,000 horas-trabajador de mano de obra se utilizan cada d√≠a. Estime el cambio en la producci√≥n que resultar√° si la inversi√≥n de capital se incrementa en $\$1,000$ y la mano de obra se incrementa en 2 horas-trabajador.

```python
from fractions import Fraction
K, L = symbols("K,L", nonnegative=True)
Q = 60*K**(Fraction(1,2))*L**(Fraction(1,3))
Q
```

```python
Delta_K = 1
Delta_L = 2

level = {K:900, L:1000}

Delta_Q = Q.diff(K).subs(level)*Delta_K+Q.diff(L).subs(level)*Delta_L
Delta_Q 
```
